```python
def calc_rmseSpatial(input1,input2,iRange,jRange):
    err2 = (input1-input2)**2
    err=err2**(1/2)
    output = indexMeanMetric(err,iRange,jRange)
    return err,output

# define function for calculating spatial mean
def indexMeanMetric(input,iRange,jRange):
    iRange=range(iRange[0],iRange[1]+1)
    jRange=range(jRange[0],jRange[1]+1)
    output = input.isel(eta_rho=iRange, xi_rho=jRange).mean(dim='eta_rho', skipna=True).mean(dim='xi_rho', skipna=True)
    return output
```

```python
# set up spatial averaging boxes, with names for each box
plt_params = {'linewidth':2}
rangesI1 = (115, 100, 40, 200)
rangesI2 = (200, 113, 80, 207)
rangesJ1 = (85, 40, 150, 250)
rangesJ2 = (150, 80, 200, 270)
names = ('no1','no2','no3','no4')


# loop through names and ranges, and add each into an xarray of all outputs
for name, rngX1, rngX2, rngY1, rngY2 in zip(names, rangesI1, rangesI2, rangesJ1, rangesJ2):
    rngX=[rngX1, rngX2]
    rngY=[rngY1, rngY2]
    if name =='no1': # awkward behaviour for first loop value
        ssh_metrics = indexMeanMetric(ssh,rngX,rngY).to_dataset()
        ssh_metrics[name] = ssh_metrics['zeta']
        ssh_metrics = ssh_metrics.drop(['zeta'])
    else:           # more general behaviour once the xarray is established
        ssh_metrics = ssh_metrics.assign({name:indexMeanMetric(ssh,rngX,rngY)})






err,ssh_rmse_metrics = calc_rmseSpatial(ssh,sst.temp,rngX,rngY)
```

```python
def simple_TimeSeries(inputDataX, inputDataY, plt_kwargs={}, ax=None):
    #Plotting
    if ax is None:
        ax = plt.gca()   
    hOut = ax.plot(inputDataX, inputDataY, **plt_kwargs)
    return(hOut)


names = ('no1','no2','no3','no4')

plt_params = {'linewidth':2}


fig, axes = plt.subplots(4, figsize=(6, 8))
for ax, run in zip(axes, names):
    simple_TimeSeries(ssh_metrics['ocean_time'],ssh_metrics[run],ax=ax, plt_kwargs=plt_params)
```
